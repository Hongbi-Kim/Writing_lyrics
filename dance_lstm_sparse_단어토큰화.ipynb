# 훈련데이터 생성
- 줄바꿈 문자 기준으로 문장 토큰화

```python
df_text = df['lyric'].str.split(' ')
df_text
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts(df_text)
vocab_size = len(tokenizer.word_index) + 1
print('단어 집합의 크기 : %d' % vocab_size)
```

```python
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(df_text)
print(word_index)
print(sequences[:100])
```

```python
sequences = list()
for line in dance_text: # 줄바꿈 문자를 기준으로 문장 토큰화
    encoded = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        sequences.append(sequence)

print('학습에 사용할 샘플의 개수: %d' % len(sequences))
```

```python
max_len = max(len(l) for l in sequences) # 모든 샘플에서 길이가 가장 긴 샘플의 길이 출력
print('샘플의 최대 길이 : {}'.format(max_len))
```

- pad_sequences()를 사용해 모든 샘플의 길이 맞춰주기
- padding의 'pre'는 0값의 위치 설정 인자
 * 이 경우 샘플 최대 길이인 406보다 짧은 샘플의 앞에 0을 채워 전체 훈련데이터의 길이를 맞춤

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

sequences = pad_sequences(sequences, maxlen=max_len, padding='pre')
```

```python
sequences = np.array(sequences)
X = sequences[:,:-1]
y = sequences[:,-1]
```

## 모델 설계

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN
```

- 하이퍼파라미터 설명
 * 임베딩 벡터 차원 : 10
 * 은닉상태 크기 : 32
 * 다 대 일 구조의 RNN 사용

- 해당 문제 해결 방법 : 다중 클래스 분류 문제
 * 마지막 시점에 모든 가능한 단어 중 하나의 단어를 예측하는 문제로 접근
 * 활성화 함수: 소프트맥스 함수
 * 손실 함수: 크로스 엔트로피 함수
 * 에포크 : 200

```python
from tensorflow.keras.layers import LSTM
```

```python
embedding_dim = 10
hidden_units = 128

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(LSTM(hidden_units))
model.add(Dense(vocab_size, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=2)
```

```python
def sentence_generation(model, tokenizer, current_word, n): # 모델, 토크나이저, 현재 단어, 반복할 횟수
    init_word = current_word
    sentence = ''

    # n번 반복
    for _ in range(n):
        encoded = tokenizer.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen=max_len-1, padding='pre')

        # 입력한 X(현재 단어)에 대해서 y를 예측하고 y(예측한 단어)를 result에 저장.
        result = model.predict(encoded, verbose=0)
        result = np.argmax(result, axis=1)

        for word, index in tokenizer.word_index.items(): 
            # 만약 예측한 단어와 인덱스와 동일한 단어가 있다면
            if index == result:
                break
        #else:

        # 현재 단어 + ' ' + 예측 단어를 현재 단어로 변경
        current_word = current_word + '\n'  + word

        # 예측 단어를 문장에 저장
        sentence = sentence + '\n' + word

    sentence = init_word + sentence
    return sentence
```
