```python
import pandas as pd
from scipy.stats import bartlett
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/writing_lyrics/assessment.csv')
```

```python
df = df[['댄스_가사1(평가)', '댄스_가사2(평가)', '댄스_가사3(평가)', '발라드_가사1(평가)','발라드_가사2(평가)', '발라드_가사3(평가)']][:-2]
df
```

- 댄스

```python
# H0: 등분산이다.
# Ha : 등분산이 아니다
bartlett(df['댄스_가사1(평가)'], df['댄스_가사2(평가)'], df['댄스_가사3(평가)'])
```

```python
bartlett(df['발라드_가사1(평가)'], df['발라드_가사2(평가)'], df['발라드_가사3(평가)'])
```

```python
import scipy.stats as stats
```

```python
# HO : 평균1 = 평균2 = 평균3
# HA : 귀무가설이 아니다.
f_statistic, p_val = stats.f_oneway(df['댄스_가사1(평가)'], df['댄스_가사2(평가)'], df['댄스_가사3(평가)'])
print('f_statistic : ', f_statistic)
print('p_val : ', p_val)
# ANOVA는 평균 차이 검정
# p-value가 유의하지 않다.(>0.5) 귀무가설 기각을 못하니까 -> 평균이 모두 같다. -> 
```

```python
f_statistic, p_val = stats.f_oneway(df['발라드_가사1(평가)'], df['발라드_가사2(평가)'], df['발라드_가사3(평가)'])
print('f_statistic : ', f_statistic)
print('p_val : ', p_val)
```

```python
# t - test
from scipy import stats
stats.ttest_ind(df['댄스_가사1(평가)'], (df['댄스_가사2(평가)']+df['댄스_가사3(평가)'])/2)
# 댄스 : 1 vs 2
# H0 : 평균1 = 평균2
# Ha : 평균1 != 평균2
```
